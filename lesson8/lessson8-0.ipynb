## 基本 sql passing by value

```python


import pandas as pd
# 連線到 raspberry 
import psycopg2
from dotenv import load_dotenv, find_dotenv
import os
import sys
success = load_dotenv()

if success:
    #print("成功載入 .env 檔案\n")
    pass
else:
    load_dotenv(find_dotenv(f'.env.raspberry'))
    #load_dotenv(find_dotenv(f'.env.render'))

#把敏感資料改成從 .env檔案 讀取
conn = psycopg2.connect(host     = os.getenv('POSTGRE_HOST'),
                        database = os.getenv('POSTGRE_DATABASE'),
                        user     = os.getenv('POSTGRE_USER'),
                        password = os.getenv('POSTGRE_PASSWORD'))

SQL_all_data_TWII =  '''
SELECT country,市場.name,date,adj_close,volume
FROM 股市 JOIN 市場 ON 股市.name = 市場.name
WHERE country = %s;
'''
    
def getData(_country):
    with conn:
        with conn.cursor() as cursor:
            cursor.execute(SQL_all_data_TWII,(_country,))
            all_data = cursor.fetchall()
    conn.close()
    return all_data
#pd.DataFrame(getData())
pd.DataFrame(getData('香港恆生'), columns=['國家','代碼','日期','收盤價','成交量'])



```

## 進階 sql passing by List , 但實際上要為 tuple (value1,vlues2,...)
### 參考 SQL  的 IN 語法

```python


import pandas as pd
# 連線到 raspberry 
import psycopg2
from dotenv import load_dotenv, find_dotenv
import os
import sys
success = load_dotenv()

if success:
    #print("成功載入 .env 檔案\n")
    pass
else:
    load_dotenv(find_dotenv(f'.env.raspberry'))
    #load_dotenv(find_dotenv(f'.env.render'))

#把敏感資料改成從 .env檔案 讀取
conn = psycopg2.connect(host     = os.getenv('POSTGRE_HOST'),
                        database = os.getenv('POSTGRE_DATABASE'),
                        user     = os.getenv('POSTGRE_USER'),
                        password = os.getenv('POSTGRE_PASSWORD'))

SQL_all_data_TWII =  '''
SELECT country,市場.name,date,adj_close,volume
FROM 股市 JOIN 市場 ON 股市.name = 市場.name
WHERE country IN %s;
'''
    
def getData(_country:tuple[str])->list[tuple]:
    with conn:
        with conn.cursor() as cursor:
            cursor.execute(SQL_all_data_TWII,(_country,))
            all_data = cursor.fetchall()
    conn.close()
    return all_data
#pd.DataFrame(getData())
pd.DataFrame(getData(('台灣','香港恆生')),columns=['國家','代碼','日期','收盤價','成交量'])

```
